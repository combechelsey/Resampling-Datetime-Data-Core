```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter, YearLocator, MonthLocator


# Load Data
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"

df = pd.read_csv(url)
df
```

```python
# Convert the date column to datetime dtype
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['date']
```

```python
# Set the date column as the index.
df = df.set_index('date')
```

- Keep only data from the year 2000 or later.

- Keep only the following features:
        precipitation
        mean_temp
        min_temp
        max_temp
        snow_depth


```python
# Keep only information listed above  

df = df.loc['2000':, ['precipitation', 'mean_temp', 'min_temp', 
                      'max_temp', 'snow_depth']]
df

```

- Impute any missing values.

    Think about each feature and what imputation method makes sense. You may not be able to use the same method for every feature!


```python
# Check for missing values 
df.isna().sum()
```

```python
# Interpolate missing data for mean, min, and max temp categories 

df['mean_temp'] = df['mean_temp'].interpolate()
df['min_temp'] = df['min_temp'].interpolate()
df['max_temp'] = df['max_temp'].interpolate()
```

```python
# Checking that null values were imputed
df.isna().sum()
```

```python
# Impute values of 0 for precipitation and snow depth

df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
```

```python
# Checking that null values were imputed with 0
df.isna().sum()
```

**Q1: What month had the most precipitation between 2000 through 2010?**

- Resample the data as Monthly frequency.
- Find the date and corresponding value for the max precipitation.
- Plot the feature and time period that corresponds to the question.
    - Make sure to meet the "Visualization Requirements" below!
- Determine the date and value for the most precipitation (the max precipitation).
- Add a vertical line at this date with a label that says the date (using date format: "April 2000") and the max precipitation value.
    - Make sure to add a legend so that your label is visible.


```python
# Making a copy of imputed dataframe
df1 = df.copy()
```

```python
# Keep data from 2000 to 2010 

df1 = df1.loc['2000':'2010']
df1
```

```python
# Saving 2000-2010 data as dataframe

q1 = df1.copy()
```

```python
# Calculate monthly precipitation totals 
q1_sum = q1['precipitation'].resample("M").sum()

print(q1_sum)
```

```python
# View info about monthy precipitation totals
q1_sum.describe()
```

```python
#import tick customization tools
import matplotlib.ticker as mticks
import matplotlib.dates as mdates

## creating our tick locators and formatters
## for minor month ticks
# create the locator to place ticks every 3 months.
loc_3months = mdates.MonthLocator(interval=3)
fmt_months = mdates.DateFormatter("%b")
## for major year ticks
YearLocator = mdates.YearLocator()
fmt_year = mdates.DateFormatter("%Y")

```

```python
month_max = q1_sum.idxmax()
month_max_date = month_max.strftime('%B %Y')
print(f"The highest monthly precipitation totals from 2000 - 2010 occurred in {month_max_date} with {q1_sum.max()} mm of total rainfall")
```

```python
fig, ax = plt.subplots(figsize=(15, 4))

# Set labels and title
ax.set_xlabel("Date")
ax.set_ylabel("Precipitation Totals (mm)")
ax.set_title("Monthly Precipitation Totals, 2000 - 2010")



# Customize the major x-axis ticks
loc_year = mdates.YearLocator(base=1)  # Major ticks every 1 year
fmt_year = mdates.DateFormatter('%Y')  # Format as 4-digit years

## customize major ticks
ax.xaxis.set_major_locator(loc_year)
ax.xaxis.set_major_formatter(fmt_year)

## customize minor ticks
ax.xaxis.set_minor_locator(loc_3months)
ax.xaxis.set_minor_formatter(fmt_months)

## Making major/minor gridlines visually distinct
ax.grid(which='minor',axis='x',ls=":")
ax.grid(which='major',axis='x',color='k')
fig.autofmt_xdate(rotation=90 ,ha='center')

ax.plot(q1_sum.index, q1_sum)

# Remove the labels for minor ticks
ax.tick_params(axis='x', which='minor', labelbottom=False)

# Add a red vertical line at the maximum value
ax.axvline(month_max, color='red', linestyle='-', label= month_max_date)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show(); 


```

**Q2: Which year between 2000-2020 had the coolest average temperature?**

- Resample the data as Yearly frequency.
- Plot the feature and time period that corresponds to the question.
    - Make sure to meet the "Visualization Requirements" below!
- Determine the date and value for the lowest mean temp.

- Add a vertical line at this date with a label that says the date (using date format: "2000") and the lowest mean temperature value.




```python
# Making a copy of imputed dataframe
df2 = df.copy()
```

```python
# Keep data from 2000 to 2010 

df2 = df2.loc['2000':'2020']
df2
```

```python
# Saving 2000-2020 data as dataframe

q2 = df2.copy()
```

```python
# Calculate yearly temperature average 
q2_ave = q2['mean_temp'].resample("Y").mean()

print(q2_ave)
```

```python
# View info about yearly temperature averages
q2_ave.describe()
```

```python
q2_ave.loc["2010"]
```

```python
# FInd Year with lowest average temperature
temp_min_date = q2_ave.idxmin()
temp_min_date.year
```

```python
# Print statement
min_temp_year = temp_min_date.year
min_mean_temp = round(q2_ave.loc["2010"].values[0], 2)


print(f"From 2000 to 2020, the year with the coolest average temperature was {min_temp_year} with an average temperature of {min_mean_temp} degrees Celsius.")
```

```python
fig, ax = plt.subplots(figsize=(12, 4))

# Set labels and title
ax.set_xlabel("Date")
ax.set_ylabel("Temperature (C)")
ax.set_title("Average Yearly Temperature, 2000 - 2020")



# Customize the major x-axis ticks
loc_year = mdates.YearLocator(base=1)  # Major ticks every 1 year
fmt_year = mdates.DateFormatter('%Y')  # Format as 4-digit years

## customize major ticks
ax.xaxis.set_major_locator(loc_year)
ax.xaxis.set_major_formatter(fmt_year)

## customize minor ticks
ax.xaxis.set_minor_locator(loc_3months)
ax.xaxis.set_minor_formatter(fmt_months)

## Making major/minor gridlines visually distinct
ax.grid(which='minor',axis='x',ls=":")
ax.grid(which='major',axis='x',color='k')
fig.autofmt_xdate(rotation=90 ,ha='center')

ax.plot(q2_ave.index, q2_ave)

# Remove the labels for minor ticks
ax.tick_params(axis='x', which='minor', labelbottom=False)

# Add a red vertical line at the maximum value
ax.axvline(temp_min_date, color='red', linestyle='-', label= {min_temp_year, min_mean_temp})
ax.legend()

# Show the plot
plt.tight_layout()
plt.show(); 
```
